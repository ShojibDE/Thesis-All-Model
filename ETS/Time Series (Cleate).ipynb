# ETS

```python


from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt

```

```python

# Simple Exponential Smoothing Method --- with Alpha = 0.20
fit1 = SimpleExpSmoothing(ts).fit(smoothing_level=0.2,optimized=False)

```

```python
# Fitted Values
fit1.fittedvalues
```

```python
# Model Parameters 
fit1.params
```

```python


# Model Parameters (Formated form)
fit1.params_formatted

```

```python


# Model Selection Criteria (AIC, AICc, BIC)
fit1.aic, fit1.aicc, fit1.bic
```

```python
# Predction for next Three Years
fcast1 = fit1.forecast(3)
fcast1
```

```python
# Simple Exponential Smoothing Method --- with Alpha = 0.20
fit2 = SimpleExpSmoothing(ts).fit(optimized=True)
```

```python
# Model Selection Criteria (AIC, AICc, BIC)
fit2.aic, fit1.aicc, fit1.bic
```
