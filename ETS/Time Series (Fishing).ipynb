# train-test split


```python
# train data
ts_train=ts.iloc[0:800]
# test data
ts_test=ts.iloc[800:]
```

```python
# number of obserbation in train data
len(ts_train)
```

```python
# number of obserbation in test data
len(ts_test)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
```

```python
# https://www.statsmodels.org/dev/examples/notebooks/generated/exponential_smoothing.html
# Simple Exponential Smoothing Method 
fit1 = SimpleExpSmoothing(ts_train).fit(optimized=True)
```

```python
# Fitted Values
fit1.fittedvalues
```

```python
# Model Parameters 
fit1.params
```

```python
# Model Parameters (Formated form)
fit1.params_formatted
```

```python
# Model Selection Criteria (AIC, AICc, BIC)
#fit1.aic, fit1.aicc, fit1.bic
```

```python
fcast1 = fit1.forecast(206)
fcast1
```

# Model Performance

```python
# Casting into  list
fitted_values= list(fit1.fittedvalues)
```

```python
# Casting into  list
ts_train= list(ts_train)
```

```python
# Casting into  list  
forecast_values=list(fcast1)
```

```python
# Casting into  list  
actual_values=list(ts_test)
```

#### Mean square error(MSE)
https://www.educative.io/answers/calculating-mean-squared-error-in-python

```python
# MSE for fitted values  values



summation = 0  #variable to store the summation of differences
n = len(actual_values) #finding total number of items in list
for i in range (0,n):  #looping through each element of the list
  difference = ts_train[i] - fitted_values[i]  #finding the difference between observed and predicted value
  squared_difference = difference**2  #taking square of the differene 
  summation = summation + squared_difference  #taking a sum of all the differences
MSE_train = summation/n  #dividing summation by total values to obtain average
print("The Mean Square Error is: " , MSE_train) 
```

```python
# MSE for forecast values



summation = 0  #variable to store the summation of differences
n = len(actual_values) #finding total number of items in list
for i in range (0,n):  #looping through each element of the list
  difference = actual_values[i] - forecast_values[i]  #finding the difference between observed and predicted value
  squared_difference = difference**2  #taking square of the differene 
  summation = summation + squared_difference  #taking a sum of all the differences
MSE_test = summation/n  #dividing summation by total values to obtain average
print("The Mean Square Error is: " , MSE_test) 
```

#### Root Mean square error(RMSE)

```python
# RMSE of train 
print("The Root Mean Square Error for train  is: " ,np.sqrt(MSE_train))
```

```python
# RMSE of test
print("The Mean Square Errorfor test is: " ,np.sqrt(MSE_test))
```

# R2
https://www.askpython.com/python/coefficient-of-determination

```python
# r2 for train data model 1
#actual = [1,2,3,4,5]
#predict = [1,2.5,3,4.9,4.9]
 
corr_matrix = np.corrcoef(ts_train, fitted_values)
corr = corr_matrix[0,1]
R_sq = corr**2

print(R_sq)
```

```python
# r2 for test data model 1
#actual = [1,2,3,4,5]
#predict = [1,2.5,3,4.9,4.9]
 
corr_matrix = np.corrcoef(actual_values, forecast_values)
corr = corr_matrix[0,1]
R_sq = corr**2
 
print(R_sq)
```

```python
# Models parameters and Model selection
# SSE = Sum square error
# minimum AIC BIC give the better result
results=pd.DataFrame(index=[r"$\alpha$",r"$\beta$",r"$\phi$",r"$\gamma$",r"$l_0$","$b_0$","SSE", "AIC","BIC"])
params = ['smoothing_level', 'smoothing_trend', 'damping_trend', 'smoothing_seasonal', 'initial_level', 'initial_trend']


results["Model-1"] = [fit1.params[p] for p in params] + [fit1.sse] + [fit1.aic] + [fit1.bic]
results["Model-2"] = [fit2.params[p] for p in params] + [fit2.sse] + [fit2.aic] + [fit2.bic]
results["Model-3"] = [fit3.params[p] for p in params] + [fit3.sse] + [fit3.aic] + [fit3.bic]
results["Model-4"] = [fit4.params[p] for p in params] + [fit4.sse] + [fit4.aic] + [fit4.bic]
results["Model-5"] = [fit5.params[p] for p in params] + [fit5.sse] + [fit5.aic] + [fit5.bic]
results["Model-6"] = [fit6.params[p] for p in params] + [fit6.sse] + [fit6.aic] + [fit6.bic]
results
```
